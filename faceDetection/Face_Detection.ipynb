```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("group_photo.jpg")
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
import cv2
```

```python
model = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces = model.detectMultiScale(img)
```

```python
faces
```

```python
faces.shape
```

```python
for face in faces:
  x,y,w,h = face

  img = cv2.rectangle(img, (x,y), (x+w, y+h), (50,205,50), 2 )
```

```python
plt.imshow(img)
```
